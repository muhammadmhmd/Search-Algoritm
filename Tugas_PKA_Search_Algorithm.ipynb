```python
import pandas as pd
```

```python
romania = pd.read_csv('Romania.csv',delimiter = ";",names = ['Cities','Neighbors','City to Neighbor Distance','City to Bucharest Distance'],header = None)
romania
```

```python
def greedy_sa(start,end,df):
  path = [start]
  traveled_distance = 0

  print('Start city:',start)
  print('End city:',end,'\n')
  n = 1

  # Greedy Algoritm
  while path[-1] != end:

    cities_neighbors = df[df['Neighbors'] == path[-1]].sort_values('City to Bucharest Distance').reset_index(drop = True)
    j = 0
    while cities_neighbors['Cities'][j] in path:
      j += 1
    nearest_city = cities_neighbors['Cities'][j]
    traveled_distance += cities_neighbors['City to Neighbor Distance'][j]

    # Neighbors and Selection Visualization
    space_max = len(max(cities_neighbors['Cities'], key=len))
    print(n,path[-1],'Heuristic Neighbors:')
    for i in range(len(cities_neighbors['Cities'])):
      if i == 0:
        print('-',cities_neighbors['Cities'][i],
              ' '*(space_max-len(cities_neighbors['Cities'][i])),':',
              cities_neighbors['City to Bucharest Distance'][i],'<-- Sortest')
      else:
        print('-',cities_neighbors['Cities'][i],
              ' '*(space_max-len(cities_neighbors['Cities'][i])),':',
              cities_neighbors['City to Bucharest Distance'][i])

    path.append(nearest_city)
    print('>> Path:',path,'\n')
    n += 1
  
  print('Final Path:',path)
  print('Total Distance:',traveled_distance)


  return path
```

```python
greedy_path = greedy_sa('Arad','Bucharest',romania)
```

```python
def A_star_sa(start,end,df):
  path = [start]
  traveled_distance = 0

  print('Start city:',start)
  print('End city:',end,'\n')
  print('f(n) = h(n) (Heuristic) + g(n) (Neighbors Distance + Traveled Distance) \n')
  n = 1

  # A* Algoritm
  while path[-1] != end:
    
    # Mencari tetangga dari elemen terakhir pada path
    cities_neighbors = romania[romania['Neighbors'] == path[-1]].sort_values('City to Bucharest Distance')
    # Menghitung fungsi evaluasi A*
    cities_neighbors['f'] = cities_neighbors['City to Neighbor Distance'] + cities_neighbors['City to Bucharest Distance'] + traveled_distance
    # Memilih tetangga dengan nilai evaluasi terkecil dan belum pernah dilalui
    cities_neighbors = cities_neighbors.sort_values('f').reset_index(drop = True)
    j = 0
    while cities_neighbors['Cities'][j] in path:
      j += 1
    nearest_city = cities_neighbors['Cities'][j]
    traveled_distance += cities_neighbors['City to Neighbor Distance'][j]

    # Neighbors and Selection Visualization
    space_max = len(max(cities_neighbors['Cities'], key=len))
    print(n,path[-1],'Neighbors f(n):')
    for i in range(len(cities_neighbors['Cities'])):
      if i == 0:
        print('-',cities_neighbors['Cities'][i],
              ' '*(space_max-len(cities_neighbors['Cities'][i])),':',
              cities_neighbors['City to Bucharest Distance'][i],'+',
              cities_neighbors['City to Neighbor Distance'][i],'+',
              traveled_distance,'=',
              cities_neighbors['f'][i],'<-- Sortest')
      else:
        print('-',cities_neighbors['Cities'][i],
              ' '*(space_max-len(cities_neighbors['Cities'][i])),':',
              cities_neighbors['City to Bucharest Distance'][i],'+',
              cities_neighbors['City to Neighbor Distance'][i],'+',
              traveled_distance,'=',
              cities_neighbors['f'][i])

    # Memasukan kota terdekat ke dalam path
    path.append(nearest_city)
    print('>> Path:',path,'\n')
    n += 1
  
  print('Final Path:',path)
  print('Total Distance:',traveled_distance)
  return path
```
